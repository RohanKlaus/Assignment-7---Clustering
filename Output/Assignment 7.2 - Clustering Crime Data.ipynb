```python
def dbscan(X,eps,min_samples):
    ss=StandardScaler()
    X=ss.fit_transform(X)
    db=DBSCAN(eps=eps,min_samples=min_samples)
    db.fit(X)
    y_pred=db.fit_predict(X)
    plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='Paired')
    plt.title("DBSCAN")
```

```python
dbscan(array_DBSCAN,1.0,8)
```

# K-means:

```python
df1 = df.drop(['Cluster'], axis=1)
df1
```

### Normalisation of Data:

```python
#Similar to the sklearn.preprocessing, StandardScaler function

def norm_func(i):
    x=(i-i.mean())/(i.std())
    return(x)
```

```python
x=norm_func(df1.iloc[:,0:])
x
```

```python
#Normalizaton Function

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

scaled_df1 = scaler.fit_transform(df1.iloc[:,0:])
```

```python
scaled_df1
```

```python
#wcss = Within Clusters Sum of Squares

wcss = []
for i in range(2,10):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_df1)
    wcss.append(kmeans.inertia_)
    
print(wcss)

plt.plot(range(2,10),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### Normalisation of Data:

```python
#Normalization Function : (MinMax Function)

def minmax_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
#Normalized data frame (considering the numerical part of data)

df_norm = minmax_func(df2.iloc[:,0:])
```

```python
df_norm
```

```python
#create dendrogram

dendrogram = sch.dendrogram(sch.linkage(df_norm,method='complete'))

#methods : 'single','complete','centroid','average'
```

```python
#create clusters

hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
#save clusters for chart

y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
Clusters = pd.DataFrame(y_hc, columns=['Clusters'])
```
